```python
import numpy as np
from math import exp
import matplotlib.pyplot as plt
%matplotlib notebook

def sigmoid(x):
    a = []
    for var in x:
        a.append(1/(1+exp(-var)))
    return a

def sigmoid_print(x):
    for var in x:
        temp = (1/(1+exp(var)))
        print("Logistic function at x = %d equals %f" % (var, temp))
    
# plot function
x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)
plt.plot(x,sig)
plt.show()

# print values
x = np.arange(-5., 6., 1.)
sigmoid_print(x)
```
